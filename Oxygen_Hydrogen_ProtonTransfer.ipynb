```python
###將n_6.xyz所有的O-H (on phenol)距離都算出來，
### 做成一個分佈圖，x軸是O-H距離，y軸是出現的次數，
###gap的中心點求得的cut-off distance，用以區分是否有proton transfer。
###Lennard Jones pairwise potential，從已經最佳化過的結構裡找出cut-off distance.

with open('Oxygen_hyrdrogen_distance.txt', 'r') as f:
    lines =f.readlines()
    # Create Data
    x = [float(line.split()[0]) for line in lines]
    y = [float(line.split()[1]) for line in lines]
    
    colors = (0,0,0)
    area = np.pi*3
    
# Create plot    
plt.scatter(x, y, s=area, c=np.array([colors]), alpha=0.5)
plt.title("Proton Transder/ Non Proton Transfer")
plt.xlabel("Oxygen and Hydrogen Distance Å")
plt.ylabel("Occurances Oxygen and Hydrogen")
plt.show()
```
