```python
### write for loop to run through phenolMeNH3+.txt file

i = 0
f = open('phenolMeNH3+.txt', 'r')

lines = f.readlines()
for lines in f:
    oxygen = lines.split()
    (b1,c1,d1) = oxygen
    x1 = float(b1)
    y1 = float(c1)
    z1 = float(d1)

    hydrogen = lines[1].split()
    (b2,c2,d2) = hydrogen
    x2 = float(b2)
    y2 = float(c2)
    z2 = float(d2)

    
    distance = math.sqrt((x1-x2)**2 + (y1-y2)**2 + (z1-z2)**2)

print(distance)
    
    
f.close()
```
